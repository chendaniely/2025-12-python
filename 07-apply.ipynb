```python
import numpy as np

def avg_2_mod(x, y):
    if (x == 20):
        return np.nan
    else:
        return (x + y) / 2
```

```python
avg_2_mod(20, 10)
```

```python
avg_2_mod(0, 10)
```

```python
assert avg_2_mod(0, 10) == 5
```

```python
import numpy as np

def avg_2_mod(x, y):
    if (x == 20):
        return np.nan
    else:
        return (x + y) / 2
```

```python
avg_2_mod_np = np.vectorize(avg_2_mod)
```

```python
avg_2_mod_np(20, 10)
```

```python
avg_2_mod_np(0, 10)
```

```python
df
```

```python
avg_2_mod_np(df.a, df.b)
```

```python
@np.vectorize
def avg_2_mod_np(x, y):
    if (x == 20):
        return np.nan
    else:
        return (x + y) / 2
```

```python
avg_2_mod_np(df.a, df.b)
```

```python
import numba
```

```python
@numba.vectorize
def avg_2_mod_numba(x, y):
    if (x == 20):
        return np.nan
    else:
        return (x + y) / 2
```

```python
avg_2_mod_numba(10, 20)
```

```python
avg_2_mod_numba(0, 10)
```

```python
avg_2_mod_numba(df.a, df.b)
```
