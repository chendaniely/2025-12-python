```python
weather = pd.read_csv("data/weather.csv")
```

```python
weather.head()
```

```python
weather_long = weather.melt(
    id_vars=["id", "year", "month", "element"],
    var_name="day",
    value_name="temp"
)
```

```python
weather_long
```

```python
weather_tidy = weather_long.pivot_table(
    index=["id", "year", "month", "day"],
    columns="element",
    values="temp"
)
weather_tidy
```

```python
weather_tidy.reset_index()
```

```python
weather_tidy.index
```

```python
weather_tidy = weather.melt(
    id_vars=["id", "year", "month", "element"],
    var_name="day",
    value_name="temp"
).pivot_table(
    index=["id", "year", "month", "day"],
    columns="element",
    values="temp"
).reset_index()

weather_tidy
```

```python
weather_tidy = (
    weather
    .melt(
        id_vars=["id", "year", "month", "element"],
        var_name="day",
        value_name="temp")
    .pivot_table(
        index=["id", "year", "month", "day"],
        columns="element",
        values="temp")
    .reset_index()
)

weather_tidy
```

```python
ebola_tidy = (ebola_long
    .drop(columns='cd_country')
    .pivot_table(
        index=["Date", "Day", "country"],
        columns="case_death",
        values="count"
    )
    .reset_index()
)
ebola_tidy
```
